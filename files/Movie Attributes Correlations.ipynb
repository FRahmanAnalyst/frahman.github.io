```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
#Read the data

df = pd.read_csv(r'E:\Desktop\Portfolio Projects\4 Movie Industry\Data\movies.csv')
```

```python
df.head(6)
```

```python
#Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(col,'-',pct_missing,'%')
```

```python
#Check column data types

df.dtypes
```

```python
#Change datatype of column 'budget', 'gross' 

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64') 
```

```python
df.dtypes
```

```python
#Extract year from release date

df['yearcorrect'] = df['released'].astype(str).str[:4]
```

```python
#Sort by gross (descending)


df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#To display all data

#pd.set_option('display.max_rows', None)
```

```python
#Check for distinct

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter plot for budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget in 100 Million Dollars')
plt.ylabel('Gross Revenue in 100 Million Dollars')

plt.show()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Plot budget vs gross revenue in seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Plot budget vs score in seaborn

sns.regplot(x='score', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
sns.heatmap(corr_matrix, annot=True)

plt.title('Correlation Metrics for Film Numeric Attributes')
plt.xlabel('Numeric Attributes')
plt.ylabel('Numeric Attributes')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
df
```

```python
#Correlation heatmap for all attributes
#corr_matrix = df.corr()

sns.heatmap(df_numerized.corr(), annot=True)

plt.title('Correlation Metrics for Film Numeric Attributes')
plt.xlabel('Numeric Attributes')
plt.ylabel('Numeric Attributes')

#plt.show()
```

```python
#Correlation numbers unstacked

corr_matrix_all = df_numerized.corr()

corr_pairs_all = corr_matrix_all.unstack()

corr_pairs_all
```

```python
#Correlation numbers unstacked and sorted

corr_pairs_all_sorted = corr_pairs_all.sort_values(ascending=False)

corr_pairs_all_sorted.head(30)
```

```python
#Extracting highly correlated pairs

high_corr = corr_pairs_all_sorted[(corr_pairs_all_sorted)>0.5 ]

high_corr
```
